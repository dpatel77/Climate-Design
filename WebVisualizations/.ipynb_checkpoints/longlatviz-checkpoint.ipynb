```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
cities = "Resources/cities.csv"
cities = pd.read_csv(cities)

cities.head(15)
```

```python
cities.to_html('cities.html',index=False)
```

```python
#City Latitude vs. Max Temp

ax2 = cities.plot.scatter(x='Lat',
                      y='Max Temp')
plt.grid(which='major',axis='both', color='gray', linestyle='-', linewidth=.5)
plt.title("City Latitude vs. Max Temperature (01/05/2017)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")


plt.savefig("maxtemp.png")
plt.show();
```

```python
#City Latitude vs. Humidity
ax2 = cities.plot.scatter(x='Lat',
                      y='Humidity')
plt.grid(which='major',axis='both', color='gray', linestyle='-', linewidth=.5)
plt.title("City Latitude vs. Humidity (01/05/2017)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")

plt.savefig("humidity.png")
plt.show();
```

```python
#City Latitude vs. Cloudiness
ax2 = cities.plot.scatter(x='Lat',
                      y='Cloudiness')
plt.grid(which='major',axis='both', color='gray', linestyle='-', linewidth=.5)
plt.title("City Latitude vs. Cloudiness (01/05/2017)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")

plt.savefig("cloudiness.png")
plt.show();
```
